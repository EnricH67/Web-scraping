+ *Entro a la pàgina web del congrés dels diputats i extrec les dades a partir d'una imatge de l'hemicicle, des d'on
   puc accedir a les dades personals de cada diputat.*
   
   *I genero un Data Frame amb esl noms/càrrecs i el Territori que representa cadasun d'ells*
   

```python
res = requests.get('https://www.congreso.es/hemiciclo')
response = TextResponse(res.url, body=res.text, encoding = 'utf-8')
```

```python
response 
```

```python
diputados = []
```

```python
for a in response.xpath('//area'):
    datos = a.xpath('@onmouseover').get().split("'")
    diputados.append({
        'Nombre' : datos[5][0:],
        'Territorio' : datos[7][0:]
                
    })
```

```python
diputados
```

```python
diputados[0]
```

```python
from operator import itemgetter
```

```python
for i in diputados:
    nombres = list(map(itemgetter('Nombre'), diputados))
    
```

```python
nombres
```

```python
for i in diputados:
    territorios = list(map(itemgetter('Territorio'), diputados))
```

```python
territorios
```

```python
final_df = pd.DataFrame({'nombre/cargo' : nombres, 'territorio' : territorios})
```
